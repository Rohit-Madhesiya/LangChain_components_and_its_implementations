**Sentence Transformers on Hugging Face**

Hugging Face sentence-transformers is a Python framework for state-of-the-art sentence, text, and image embeddings.

One of the embedding models is used in the HuggingFaceEmbeddings class. We have also added an alias for SentenceTransfromerEmbeddings for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="This is a sample text document"

res=embeddings.embed_query(text)

res
```

```python
len(res)
```

```python
doc_res=embeddings.embed_documents([text,"This is not a sample text document"])

doc_res


```
